```python
def rename_col(dataframe, sep):
    dic = {}
    for each in dataframe.columns:
        column_split = each.split(sep = sep)
        if len(column_split) == 1:
            if column_split[0][0] in dataframe.columns:
                dataframe.rename(columns = {each : column_split[0][:3]}, inplace = True)
                dic[column_split[0][:3]] = each
            else:
                dataframe.rename(columns = {each : column_split[0][:3]}, inplace = True)
                dic[column_split[0][:3]] = each
        else:
            final = ''
            for a in column_split:
                final += a[0][0]
            dic[final] = each
            dataframe.rename(columns = {each : final}, inplace = True)
    return dic
```

```python
def segmentation(row):
    if row in ['CA']:
        return '01_in_CA'
    elif row in ['MT', 'ID', "WY", 'NV', 'UT', "CO", 'AZ', 'NM']:
        return '02_Pacific'
    elif row in ['MT', 'ID', 'WY', 'NV', 'UT', 'CO', 'AZ', 'NM']:
        return '03_Mountain'
    elif row in ['OK', 'AR', 'TX', 'LA']:
        return '04_West_South_Central'
    elif row in ['KY','TN', 'MS', 'AL']:
        return '05_East_South_Central'
    elif row in ['WV', 'MD', 'DE', 'DC', 'VA', 'NC', 'SC', 'GA', 'FL']:
        return '06_South_Atlantic'
    elif row in ['ND', 'SD', 'NE', 'KS', 'MN', 'IA', 'MO']:
        return '07_West_North_Central'
    elif row in  ['WI', "IL", 'IN', 'OH', 'MI']:
        return '08_East_North_Central'
    elif row in ['NJ', 'PA', 'NY']:
        return '09_Middle_Atlantic'
    elif row in ['CT', 'RI', 'MA',' VT', 'NH', 'ME']:
        return '10_New_England'
    elif row in ['03_Mountain', '02_Pacific' ,'01_in_CA']:
        return '01_West'
    elif row in ['06_South_Atlantic' , '05_East_South_Central', '04_West_South_Central']:
        return '02_South'
    elif row in ['08_East_North_Central', '07_West_North_Central']:
        return '03_Midwest'
    elif row in ['10_New_England', '09_Middle_Atlantic']:
        return '04_Northeast'

def airlineseg(row) :
    if row in ('United Airlines', 'Alaska Airlines'):
        return '01_home'
    else:
        return '02_other'
```

```python
da['subseg_des'] = da['DSA'].apply(segmentation)
da['subseg_ori'] = da['OSA'].apply(segmentation)
da['seg_des'] = da['subseg_des'].apply(segmentation)
da['seg_ori'] = da['subseg_ori'].apply(segmentation)
da['al_code'] = da['CN'].apply(airlineseg)
da
```
